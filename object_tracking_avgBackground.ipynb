```python
# Python program to illustrate
# Background subtraction using
# concept of Running Averages

# organize imports
import cv2
import numpy as np
def nothing(x):
    pass
def control(value):
    if value%2 == 0 and value>2:
        value+=1 
        return value
    elif value<3 :
        value =3
        return value
    else:
        return value
cv2.namedWindow("Taha")

cv2.createTrackbar("pixel" ,"Taha" ,1,100 ,nothing)

# capture frames from a camera
cap = cv2.VideoCapture(0)

# read the frames from the camera
_, img = cap.read()

# modify the data type
# setting to 32-bit floating point
averageValue1 = np.float32(img)

# loop runs if capturing has been initialized.
while(1):
	# reads frames from a camera
	_, img = cap.read()
	gray1 =cv2.cvtColor(averageValue1, cv2.COLOR_BGR2GRAY)
	gray2=cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

	# using the cv2.accumulateWeighted() function
	# that updates the running average
	cv2.accumulateWeighted(gray2, gray1, 0.02)
	# converting the matrix elements to absolute values
	# and converting the result to 8-bit.
	resultingFrames1 = cv2.convertScaleAbs(averageValue1)
    # subtract the images
	subtracted = cv2.subtract(img, resultingFrames1)
	gray =cv2.cvtColor(subtracted, cv2.COLOR_BGR2GRAY)
	value = cv2.getTrackbarPos("pixel" ,"Taha")

     #put the value from trackbar as input to theshold
	ret, thresh1 = cv2.threshold(gray, value, 255, cv2.THRESH_BINARY)
     #and operator with img and frame
	colored =cv2.cvtColor(thresh1, cv2.COLOR_GRAY2BGR)
	obj = cv2.bitwise_and(colored,img)

#     # Show two output windows
	# the input / original frames window
	#cv2.imshow('InputWindow', img)
    
	# the window showing output of alpha value 0.02
	cv2.imshow('averageValue1', resultingFrames1)
    
	# the window showing output of subtraction
	cv2.imshow('subtracted', subtracted)
	# the window showing output of threshold
	cv2.imshow('gray', gray)
	# the window showing output of threshold
	cv2.imshow('threshold', thresh1)
	# the window showing output of object
	cv2.imshow('object', obj)
       
	# Wait for Esc key to stop the program
	k = cv2.waitKey(30) & 0xff
	if k == 27:
		break

# Close the window
cap.release()
	
# De-allocate any associated memory usage
cv2.destroyAllWindows()
```
