```python
import cv2 
import numpy as np 

img = cv2.imread("images/tito.jpg")

gray_img = cv2.cvtColor(img ,cv2.COLOR_BGR2GRAY)

templet = cv2.imread("images/templat.jpg" ,cv2.IMREAD_GRAYSCALE)
w ,h = templet.shape[::-1]

result= cv2.matchTemplate(gray_img ,templet ,cv2.TM_CCOEFF_NORMED)

loc = np.where(result>=.99)

for lc in zip(*loc[::-1]):
    cv2.rectangle(img ,lc ,(lc[0]+w ,lc[1]+h) , (0,255,0) ,1)

cv2.imshow("result" ,result)
cv2.imshow("img" ,img)
cv2.imshow("grayImg",gray_img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
import cv2 
import numpy as np 
import matplotlib.pyplot as plt

img = cv2.imread("images/taha2.jpg")

gray_img = cv2.cvtColor(img ,cv2.COLOR_BGR2GRAY)

templet = cv2.imread("images/templat2.jpg" ,cv2.IMREAD_GRAYSCALE)
w ,h = templet.shape[::-1]

result= cv2.matchTemplate(gray_img ,templet ,cv2.TM_CCOEFF_NORMED)

loc = np.where(result>=.99)

for lc in zip(*loc[::-1]):
    cv2.rectangle(img ,lc ,(lc[0]+w ,lc[1]+h) , (0,255,0) ,3)
cv2.imwrite("output.jpg" ,img)
img = cv2.cvtColor(img , cv2.COLOR_BGR2RGB)
plt.imshow(img)
plt.show()
```
