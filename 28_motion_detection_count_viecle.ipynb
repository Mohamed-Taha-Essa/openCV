```python
import cv2
import numpy as np 

def find_center(x,y,w,h):
    x1 =int(w/2)
    y1 =int(h/2) 
    cx =x+x1
    cy =y+y1
    return cx,cy 

detect =[]
w_car = 70
h_car = 70
offset = 3
y1 = 270
car_num=0

cap =cv2.VideoCapture("Cars.mp4")

bgs =cv2.bgsegm.createBackgroundSubtractorMOG()

width = cap.get(3)
height = cap.get(4)
fps =  cap.get(cv2.CAP_PROP_FPS)
print(width)
print(height)
print(fps)
while(cap.isOpened()):
    ret ,frame=cap.read()

    gray =cv2.cvtColor(frame , cv2.COLOR_BGR2GRAY)
    blur =cv2.GaussianBlur(gray ,(5,5),0)
    img_sub = bgs.apply(blur)
#     ret ,thresh =cv2.threshold(blur , 40 ,255 ,0)
    dilate = cv2.dilate(img_sub ,np.ones((10,10)))
    
    contours ,hierarchy =cv2.findContours(dilate, cv2.RETR_TREE, cv2.CHAIN_APPROX_SIMPLE)
    cv2.line(frame ,(0,y1),(640,y1),(255,0,0),1)

    for index,contour in enumerate(contours) :
        (x,y,w,h)= cv2.boundingRect(contour)
        validat_contour =(w>w_car)and(h>h_car)
        
        if(not(validat_contour)):
            continue
        cv2.rectangle(frame,(x ,y ),(x+w ,y+h) ,(100,100,0), 2)
        center = find_center(x,y,w,h)
        detect.append(center)
        cv2.circle(frame ,center,4 ,(0,0,255) ,-1)
        
        for x ,y in detect :
            if(y < y1+offset )and (y>y1-offset):
                car_num+=1
                cv2.line(frame ,(0,y1),(640,y1),(0,0,255),3)
                detect.remove((x,y))
                print("no of cars detected:"+str(car_num))
    cv2.putText(frame , "Viecle count :"+str(car_num) ,(30,30),
                  cv2.FONT_HERSHEY_SIMPLEX , 1 ,(255,255,0),3)
        
    cv2.imshow("frame" ,frame )
    cv2.imshow("img_sub" ,img_sub )
    cv2.imshow("dilate" ,dilate )
    
    if cv2.waitKey(25)==27: 
        break
        
cv2.destroyAllWindows()
cap.release()
    
    
```
