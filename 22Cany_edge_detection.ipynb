```python
import numpy as np
import cv2 as cv
from matplotlib import pyplot as plt
img = cv.imread('salah.jpg',0)
edges = cv.Canny(img,100,200)

plt.subplot(121),plt.imshow(img,cmap = 'gray')
plt.title('Original Image'), plt.xticks([]), plt.yticks([])
plt.subplot(122),plt.imshow(edges,cmap = 'gray')
plt.title('Edge Image'), plt.xticks([]), plt.yticks([])
```

```python
import numpy as np
import cv2 as cv
def nothing(x) :
    pass

cv.namedWindow("cany")
cv.createTrackbar("start" ,"cany" , 100 ,255 ,nothing)
cv.createTrackbar("end" ,"cany" , 200 ,255 ,nothing)

img = cv.imread('salah.jpg',0)
cv.imshow("image"  ,img)
while True: 
    min_cany = cv.getTrackbarPos("start" ,"cany")
    max_cany = cv.getTrackbarPos("end" ,"cany")
    
    edges = cv.Canny(img,min_cany ,max_cany)
    
    cv.imshow("cany" ,edges)
    
    k = cv.waitKey(40)
    
    if k ==27 :
        break 
        
cv.destroyAllWindows()

    
```
